### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer, classification_report
import xgboost as xgb
from xgboost import plot_tree
```

### Import Data

```python
# !kaggle datasets download -d yeanzc/telco-customer-churn-ibm-dataset
```

```python
data = pd.read_excel('Telco_customer_churn.xlsx')
```

```python
data.head()
```

### Data preprocessing and Cleaning

```python
# remove unwanted columns
drop_cols = ['CustomerID', 'Count', 'Country', 'State', 'Lat Long', 
             'Churn Value', 'Churn Score', 'CLTV', 'Churn Reason']

data.drop(columns = drop_cols, inplace=True)

data.head()
```

```python
# replace the whitespaces with underscores in the data and columns
data.replace(' ', '_', regex=True, inplace=True)

data.columns = data.columns.str.replace(' ', '_')

data.head()
```

```python
# check for null values
data.isnull().sum()
```

```python
# check the data types
data.info()
```

```python
#  replace '_' values for column Total_Charges with 0

data.loc[data['Total_Charges'] == '_']
```

```python
data.loc[(data['Total_Charges'] == '_'), 'Total_Charges'] =0

# convert the data type to numeric
data['Total_Charges'] = pd.to_numeric(data['Total_Charges'])
```

```python
# verify the data type change for the 'Total_Charges' column
data.info()
```

### Label Encoding

```python
X = data.drop(columns = ['Churn_Label'])
y = data['Churn_Label']
```

```python
X.columns
```

```python
X_encoded = pd.get_dummies(X, columns = ['City', 'Gender', 'Senior_Citizen', 'Partner', 'Dependents', 'Phone_Service', 
                             'Multiple_Lines', 'Internet_Service', 'Online_Security', 'Online_Backup', 
                             'Device_Protection', 'Tech_Support', 'Streaming_TV', 'Streaming_Movies', 
                             'Contract', 'Paperless_Billing', 'Payment_Method'])

X_encoded.head()
```

```python
y_encoded = pd.get_dummies(y, drop_first=True)['Yes']
y_encoded.unique()
```

# Modeling

```python
# perform stratified train-test split to maintain the class ratio

print('Original Ratio: ', sum(y_encoded)/len(y_encoded))
      
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, random_state=42, stratify=y_encoded)
      
print('Spit Ratio Train: ', sum(y_train)/len(y_train))
print('Spit Ratio test: ', sum(y_test)/len(y_test))
```

```python
# function to print the evalaution metrics

def eval_metrics(model, X_test, y_test):
    y_pred_model = model.predict(X_test)
    
    conf_mat = confusion_matrix(y_test, y_pred_model)
    
    ConfusionMatrixDisplay(conf_mat, display_labels=['Did not leave', 'Left']).plot();
    
    print('Precision: ', precision_score(y_test, y_pred_model))
    print('Recall: ', recall_score(y_test, y_pred_model))
    print('F1 Score: ', f1_score(y_test, y_pred_model))
```

## 5. XGBoost

```python
# build and train the model

churn_xgb = xgb.XGBClassifier(objective='binary:logistic', seed=42)

churn_xgb.fit(X_train, y_train,
              verbose=True,
              early_stopping_rounds=10,
              eval_metric='aucpr',
              eval_set=[(X_test, y_test)]
             )
```

### Evaluating the model

```python
eval_metrics(churn_xgb, X_test, y_test)
```

- Key thing to check for in this matrix is that we want the right bottom box to be optimized as we are concerned with the customer who is leaving. We want to correctly reduce the error for "Left" even if it means increasing errors in "Did not leave". Here recall is a good metric to look at for this situation which is....
### Very BAD!!!

### Hyperparameter tuning for XGBoost

#### Round 1 - Optimizing for max_depth and learning rate

```python
grid_params_1 = {'max_depth': [3,4,5,6],
                 'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.5]
                }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
grid_1 = GridSearchCV(estimator=xgb_model, param_grid=grid_params_1, n_jobs=3, cv=3,verbose=1, scoring='roc_auc')
grid_1.fit(X_train, y_train)
```

```python
print('Best Params: ', grid_1.best_params_)
print('Best Score: ', grid_1.best_score_)
```

- Setting the learning_rate =0.05 and max_depth=4 obtained from above and optimizing for other params

#### Round 2 - Optimizing for gamma, reg_lambda, scale_pos_weight

```python
grid_params_2 = {'max_depth': [4],
                 'learning_rate': [0.05],
                 'gamma': [0, 0.25, 0.5, 0.75],
                 'reg_lambda': [0, 2, 5, 10],
                 'scale_pos_weight': [1,3,5]
                }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
grid_2 = GridSearchCV(estimator=xgb_model, param_grid=grid_params_2, n_jobs=3, cv=3,verbose=1, scoring='roc_auc')
grid_2.fit(X_train, y_train)
```

```python
print('Best Params: ', grid_2.best_params_)
print('Best Score: ', grid_2.best_score_)
```

- Setting the learning_rate =0.05, max_depth=4, scale_pos_weight=3, gamma=0 as obtained from above and optimizing for other params

#### Round 3 - Optimizing for reg_lambda again

```python
grid_params_3 = {'max_depth': [4],
                 'learning_rate': [0.05],
                 'gamma': [0],
                 'reg_lambda': [5, 10, 15, 20],
                 'scale_pos_weight': [3]
                }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
grid_3 = GridSearchCV(estimator=xgb_model, param_grid=grid_params_3, n_jobs=3, cv=3,verbose=1, scoring='roc_auc')
grid_3.fit(X_train, y_train)
```

```python
print('Best Params: ', grid_2.best_params_)
print('Best Score: ', grid_2.best_score_)
```

### Fitting the Model with optimal parameter

```python
churn_xgb_optimal = xgb.XGBClassifier(objective='binary:logistic', seed= 42,
                                      gamma=0, learning_rate=0.05, max_depth=4, 
                                      reg_lambda=10, scale_pos_weight=3
                                     )
churn_xgb_optimal.fit(X_train, y_train,
                     verbose=True,
                     early_stopping_rounds=15,
                     eval_metric='aucpr',
                     eval_set=[(X_test, y_test)])
```

```python
eval_metrics(churn_xgb_optimal, X_test, y_test)
```

As expected, there was an improvement in the recall metric. It is increased from 53.3% to 82.4%. However, This is still not great :(
