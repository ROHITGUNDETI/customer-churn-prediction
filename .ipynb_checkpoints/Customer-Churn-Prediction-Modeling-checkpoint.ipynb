### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer, classification_report

import xgboost as xgb
from xgboost import plot_tree

from sklearn.svm import SVC
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings("ignore")
```

### Import Data

```python
# !kaggle datasets download -d yeanzc/telco-customer-churn-ibm-dataset
```

```python
data = pd.read_excel('Telco_customer_churn.xlsx')
```

```python
data.head()
```

### Data preprocessing and Cleaning

```python
# remove unwanted columns
drop_cols = ['CustomerID', 'Count', 'Country', 'State', 'City', 'Zip Code', 'Lat Long', 'Latitude',
             'Longitude', 'Churn Value', 'Churn Score', 'CLTV', 'Churn Reason']

data.drop(columns = drop_cols, inplace=True)

data.head()
```

```python
# replace the whitespaces with underscores in the data and columns
data.replace(' ', '_', regex=True, inplace=True)

data.columns = data.columns.str.replace(' ', '_')

data.head()
```

```python
# check for null values
data.isnull().sum()
```

```python
# check the data types
data.info()
```

```python
#  replace '_' values for column Total_Charges with 0

data.loc[data['Total_Charges'] == '_']
```

```python
data.loc[(data['Total_Charges'] == '_'), 'Total_Charges'] =0

# convert the data type to numeric
data['Total_Charges'] = pd.to_numeric(data['Total_Charges'])
```

```python
# verify the data type change for the 'Total_Charges' column
data.info()
```

### Label Encoding

```python
# encoding gender
data['Gender'] = data['Gender'].apply(lambda x: 0 if x=='Male' else 1)

# encoding "yes-no" columns
birnay_enc_cols = ['Senior_Citizen', 'Partner', 'Dependents', 'Phone_Service', 'Multiple_Lines', 
                   'Online_Security', 'Online_Backup', 'Device_Protection', 'Tech_Support',
                  'Streaming_TV', 'Streaming_Movies', 'Paperless_Billing', 'Churn_Label']

for col in birnay_enc_cols:
    data[col] = data[col].apply(lambda x: 0 if x == "No" else 1)
```

```python
data_encoded = pd.get_dummies(data, columns = ['Internet_Service', 'Contract', 'Payment_Method'], drop_first=True)

data_encoded.head()
```

```python
X = data_encoded.drop(columns = ['Churn_Label'])
y = data_encoded['Churn_Label']
```

```python
X.head()
```

# Modeling

```python
# perform stratified train-test split to maintain the class ratio

print('Original Ratio: ', sum(y)/len(y))
      
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
      
print('Spit Ratio Train: ', sum(y_train)/len(y_train))
print('Spit Ratio test: ', sum(y_test)/len(y_test))

Models = []
```

```python
# function to print the evalaution metrics

def eval_metrics(model, X_test, y_test):
    y_pred_model = model.predict(X_test)
    
    conf_mat = confusion_matrix(y_test, y_pred_model)
    
    ConfusionMatrixDisplay(conf_mat, display_labels=['Did not leave', 'Left']).plot();
    
    print('Precision: ', precision_score(y_test, y_pred_model))
    print('Recall: ', recall_score(y_test, y_pred_model))
    print('F1 Score: ', f1_score(y_test, y_pred_model))
    print('ROC AUC Score: ', roc_auc_score(y_test, y_pred_model))
```

```python
# function to perform gridsearchcv

def gridsearchcv(model, params, scoring):
    grid = GridSearchCV(estimator=model, param_grid=params, n_jobs=4, cv=5,verbose=1, scoring=scoring)
    grid.fit(X_train, y_train)
    
    return grid
```

## 1. Logistics Regression

```python
from sklearn.linear_model import LogisticRegression

churn_logistic = LogisticRegression(random_state=42, max_iter=1000)

churn_logistic.fit(X_train, y_train)
```

```python
eval_metrics(churn_logistic, X_test, y_test)
```

### Hyperparamter tuning for L1 and L2

```python
log_parmas_1 = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

churn_logistics = LogisticRegression(random_state=42, max_iter=1000)
log_grid_1 = gridsearchcv(churn_logistics, log_parmas_1, 'roc_auc')

print('Best Params: ', log_grid_1.best_params_)
print('Best Score: ', log_grid_1.best_score_)
```

```python
churn_logistic_optimal = LogisticRegression(random_state=42, C=1, penalty='l1', solver='liblinear')

churn_logistic_optimal.fit(X_train, y_train)
Models.append(churn_logistic_optimal)
```

```python
eval_metrics(churn_logistic_optimal, X_test, y_test)
```

## 2. Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

churn_GNB = GaussianNB()

churn_GNB.fit(X_train, y_train)

Models.append(churn_GNB)
```

```python
eval_metrics(churn_GNB, X_test, y_test)
```

## 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

churn_RF = RandomForestClassifier(random_state=42)

churn_RF.fit(X_train, y_train)
```

```python
eval_metrics(churn_RF, X_test, y_test)
```

### Hyperparameter tuning for Random Forest

```python
rf_params_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [4, 5, 6, 7, 8],
    'max_features': ['sqrt', 'log2'],
    'class_weight': ['balanced', 'balanced_subsample'],
    'ccp_alpha': [0.001, 0.01, 0.1]
}

rf_grid_1 = gridsearchcv(churn_RF, rf_params_grid, 'roc_auc')

print('Best Params: ', rf_grid_1.best_params_)
print('Best Score: ', rf_grid_1.best_score_)
```

```python
best_params = rf_grid_1.best_params_
churn_rf_optimal = RandomForestClassifier(n_estimators = best_params['n_estimators'],
                                          max_depth= best_params['max_depth'],
                                          max_features= best_params['max_features'],
                                          class_weight = best_params['class_weight'],
                                          ccp_alpha= best_params['ccp_alpha'],
                                          random_state=42)
                                     
churn_rf_optimal.fit(X_train, y_train)

Models.append(churn_rf_optimal)
```

```python
eval_metrics(churn_rf_optimal, X_test, y_test)
```

```python
# understanding feature importance for this model
importances = churn_rf_optimal.feature_importances_

feature_importances = pd.DataFrame({'feature': X_train.columns, 'importance': importances})
feature_importances = feature_importances.sort_values(by='importance', ascending=True)


plt.figure(figsize=(12, 6))
plt.barh(feature_importances['feature'], feature_importances['importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')
plt.show()
```

## 5. XGBoost

```python
# build and train the model

churn_xgb = xgb.XGBClassifier(objective='binary:logistic', seed=42)

churn_xgb.fit(X_train, y_train,
              early_stopping_rounds=10,
              eval_metric='auc',
              eval_set=[(X_test, y_test)]
             )
```

```python
eval_metrics(churn_xgb, X_test, y_test)
```

- Key thing to check for in this matrix is that we want the right bottom box to be optimized as we are concerned with the customer who is leaving. We want to correctly reduce the error for "Left" even if it means increasing errors in "Did not leave". Here recall is a good metric to look at for this situation which is....
### Very BAD!!!

### Hyperparameter tuning for XGBoost

#### Round 1 - Optimizing for max_depth and learning rate and gamma(gain value threshold)

```python
grid_params_1 = {'max_depth': [3,4,5],
                 'learning_rate': [0.01, 0.05, 0.1, 0.2],
                }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
grid_1 = gridsearchcv(xgb_model, grid_params_1, 'roc_auc')

print('Best Params: ', grid_1.best_params_)
print('Best Score: ', grid_1.best_score_)
```

- Setting the learning_rate =0.05 and max_depth=4 obtained from above and optimizing for other params

#### Round 2 - Optimizing for gamma, reg_lambda, scale_pos_weight

```python
grid_params_2 = {'max_depth': [3, 4],
                 'learning_rate': [0.05, 0.1],
                 'gamma': [0, 0.25, 0.5, 0.75],
                 'reg_lambda': [0, 5, 10, 15],
                 'scale_pos_weight': [3,5,7]
                }

xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
grid_2 = gridsearchcv(xgb_model, grid_params_2, 'roc_auc')

print('Best Params: ', grid_2.best_params_)
print('Best Score: ', grid_2.best_score_)
```

### Fitting the Model with optimal parameter

```python
best_params = grid_2.best_params_
churn_xgb_optimal = xgb.XGBClassifier(objective='binary:logistic', learning_rate=best_params['learning_rate'],
                                      max_depth=best_params['max_depth'], reg_lambda=best_params['reg_lambda'],
                                      gamma=best_params['gamma'], scale_pos_weight=best_params['scale_pos_weight'],
                                      seed= 42)

churn_xgb_optimal.fit(X_train, y_train,
                     early_stopping_rounds=10,
                     eval_metric='aucpr',
                     eval_set=[(X_test, y_test)])

Models.append(churn_xgb_optimal)
```

```python
eval_metrics(churn_xgb_optimal, X_test, y_test)
```

As expected, there was an improvement in the recall metric. It is increased from 53.3% to 88% but at the cost of precision. However, This is still not great :(

```python
# understanding feature importance for this model
importances = churn_xgb_optimal.feature_importances_

feature_importances = pd.DataFrame({'feature': X_train.columns, 'importance': importances})
feature_importances = feature_importances.sort_values(by='importance', ascending=True)


plt.figure(figsize=(12, 6))
plt.barh(feature_importances['feature'], feature_importances['importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')
plt.show()
```

## 6. Kernel SVM

```python
# perform scaling on continuous data

from sklearn.preprocessing import scale

X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

```python
churn_svm = SVC(random_state=42)

churn_svm.fit(X_train_scaled, y_train)
```

```python
eval_metrics(churn_svm, X_test_scaled, y_test)
```

```python
svm_grid_params = {
    'C': [0.5, 1, 10, 20],
    'gamma': ['scale', 1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}

svm_grid = gridsearchcv(churn_svm, svm_grid_params, 'roc_auc')

print('Best Params: ', svm_grid.best_params_)
print('Best Score: ', svm_grid.best_score_)
```

```python
best_params = svm_grid.best_params_
churn_svm_optimal = SVC(C=best_params['C'], gamma=best_params['gamma'], kernel='rbf', random_state= 42)

churn_svm_optimal.fit(X_train_scaled, y_train)

Models.append(churn_svm_optimal)
```

```python
eval_metrics(churn_svm_optimal, X_test_scaled, y_test)
```

## Model Comparision

```python
from sklearn.metrics import roc_curve, roc_auc_score


plt.figure(figsize=(8, 6))

for model in Models:
    y_pred_prob = model.predict(X_test)
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    auc = roc_auc_score(y_test, y_pred_prob)
    plt.plot(fpr, tpr, label=f'{model.__class__.__name__} (AUC = {auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
